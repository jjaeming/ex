```python
import numpy as np
import matplotlib.pyplot as plt
import math

def sigmoid(x):
    return 1/(1+np.exp(-x));

degree = 2 # features of x
m = 1000 #학습 데이터 개수
half_m = int(m/2)

x1_data = np.array((np.random.normal(0.5, 0.3, half_m), np.random.normal(-0.5, 0.3, half_m)))
x1_data = np.resize(x1_data, m)

x2_data = np.array((np.random.normal(-0.5, 0.3, half_m), np.random.normal(0.5, 0.3, half_m)))
x2_data = np.resize(x2_data, m)

y_data = np.array((np.ones(half_m), np.zeros(half_m)))
y_data = np.resize(y_data, m)

theta = np.random.normal(0.0, 0.5, degree+1).T

plt.plot(x1_data[0:half_m], x2_data[0:half_m], 'r.',label = '1')
plt.plot(x1_data[half_m:], x2_data[half_m:], 'b.',label = '0')
plt.legend()
```

${\partial J(W) \over \partial W_j} = \frac{1}{m} \sum^{m}_{i=1}(\hat{y}^i-y^i)x^i_j$

```python
num_epoch = 1000
learning_rate = 0.03
errors = []
w1 = np.random.uniform(low=-1.0, high=1.0)
w2 = np.random.uniform(low=-1.0, high=1.0)

b = np.random.uniform(low=-1.0, high=1.0)

for epoch in range(num_epoch):
    
    y_hat = sigmoid(w1 * x1_data + w2*x2_data+ b)

    error = ((y_hat - y_data)**2).mean()
    if error < 0.00001:
        break

    w1 = w1 - learning_rate * ((y_hat - y_data) * x1_data).mean()
    w2 = w2 - learning_rate * ((y_hat - y_data) * x2_data).mean()
    b = b - learning_rate * (y_hat - y_data).mean()
    
    errors.append(error)
    
    if epoch % 5 == 0:
        print("{0:2} w1 = {1:.5f},b = {2:.5f}, error = {3:.5f}".format(epoch, w1, b, error))

```

```python
x_test=np.array([-0.5,0.0])
test_result = sigmoid(w1 * x_test[0] + w2*x_test[1]+ b)
print(test_result)

if test_result <0.5:
    test_class=0
elif test_result >=0.5:
    test_class=1
    
print(test_class)
```
